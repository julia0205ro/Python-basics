```python
 import pandas as pd #import pandas library
```

```python
data = pd.read_csv('sales.csv') #upload dataframe
```

```python
data.head() #display first five rows of the dataframe
```

```python
second_data = pd.read_csv('returns.csv') #upload second dataframe
```

```python
second_data.head() #display first five rows of the second dataframe
```

```python
t = data.merge(second_data) #merge two tables on common column 'order_id'
```

```python
t.head() #display first five rows of the merged dataframe
```

```python
#filter the merged table by categories 'Одежда и обувь' and 'Бытовая техника'

a = ['Одежда и обувь', 'Бытовая техника']
t_1 = t.loc[t['category'].isin(a)]
t_1.head()
```

```python
#find out a region with the biggest sales

b = t.groupby('region').sum().sales.reset_index()
b[b['sales'] == b['sales'].max()]
```

```python
#find out the return rate for each product in each region 

pivot = t.pivot_table(index='category', columns='return', values='sales', aggfunc='sum')
pivot['returns_share'] = pivot[1] / (pivot[0] + pivot[1])
pivot
```
